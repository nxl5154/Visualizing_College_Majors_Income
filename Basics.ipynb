## Visualizing Recent College Grads Salaries

The raw data for this survey was provided by the American Community Survey, and cleaned by FiveThirtyEight, in the file `recent-grads.csv`. The parameters of interest are:

- `Sample_size` - Size of student sample
- `Median` - Average income of recent graduates
- `Unemployment_rate` - Percent of unemployed graduates
- `ShareWomen` - Percent of women out of total students
- `Men` - Number of men
- `Women` - Number of women

We form a set of scatter plots to observe some of the correlations in this dataset. The interesting findings from these scatter plots are: 

- Median income is negatively correlated with sample size.
- Median income is negatively correlated with the number of full time students.
- Median income is not correlated with proportion of women in major.

Next, we form a set of histograms to see what findings we can discover. We find:

- The most common salary range is \$30,000 - \$35,000.
- 55.8% of majors are predominantly female, while 44.2% of majors are predominantly male.
